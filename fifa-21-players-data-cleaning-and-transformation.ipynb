```python
#as you can see above the data types read object and the suitable data types should be int/float
#so conversion would be

fifa_data["Height"] = pd.to_numeric(fifa_data["Height"], errors="coerce")  # Convert to int/float
fifa_data["Weight"] = pd.to_numeric(fifa_data["Weight"], errors="coerce")  # Convert to float

print(fifa_data["Height"].dtype)
print(fifa_data["Weight"].dtype)# Check updated data types
```

```python
#let us split the joined column into yy, mm, dd column

fifa_data["Loan Date End"] = pd.to_datetime(fifa_data["Loan Date End"], errors='coerce')

# Creating separate columns for Year, Month, and Day
fifa_data["Loan Year"] = fifa_data["Loan Date End"].dt.year
fifa_data["Loan Month"] = fifa_data["Loan Date End"].dt.month
fifa_data["Loan Day"] = fifa_data["Loan Date End"].dt.day



```

```python
#Cleaning and transforming the data

def convert_currency(value):
    if isinstance(value, str):
        value = value.replace("€", "").replace("K", "000").replace("M", "000000")
        try:
            return int(float(value))  # Convert to float first to handle decimals, then to int
        except ValueError:
            return np.nan  # Return NaN for invalid values
    return np.nan

# Clean and transform Value, Wage, and Release Clause columns
for col in ["Value", "Wage", "Release Clause"]:
    if col in fifa_data.columns:
        fifa_data[col] = fifa_data[col].apply(convert_currency)


# Display the first few rows of the cleaned columns
print(fifa_data[["Value", "Wage", "Release Clause"]].head())
```

```python
# Remove newline characters from the Hits column
fifa_data["Hits"] = fifa_data["Hits"].astype(str).str.replace(r'\n', '', regex=True)

# Display the first few rows to confirm the changes
print(fifa_data[["Hits"]].head())
```

```python
# Ensure the column exists before processing
if "Team & Contract" in fifa_data.columns:
    # Split the column into two based on newline (\n)
    fifa_data[["Team", "Contract Type"]] = fifa_data["Team & Contract"].str.split("\n", expand=True)

    # Display the first few rows to confirm the changes
    print(fifa_data[["Team", "Contract Type"]].head())
```
